```python
import pandas as pd

file_path = r"C:\Users\\Asus\\Desktop\\Scraper\\Scraper\\output\\25.22.04_MX_videos.csv"

data = pd.read_csv(file_path)

data.head()
```

```python
category_mapping = {
    1: "Film & Animation",
    2: "Autos & Vehicles",
    10: "Music",
    15: "Pets & Animals",
    17: "Sports",
    19: "Travel & Events",
    20: "Gaming",
    22: "People & Blogs",
    23: "Comedy",
    24: "Entertainment",
    25: "News & Politics",
    26: "Howto & Style",
    27: "Education",
    28: "Science & Technology",
    29: "Nonprofits & Activism",
    18: "Short Movies",
    21: "Videoblogging",
    30: "Movies",
    31: "Anime/Animation",
    32: "Action/Adventure",
    33: "Classics",
    34: "Comedy",
    35: "Documentary",
    36: "Drama",
    37: "Family",
    38: "Foreign",
    39: "Horror",
    40: "Sci-Fi/Fantasy",
    41: "Thriller",
    42: "Shorts",
    43: "Shows",
    44: "Trailers"
}

data['category_title'] = data['categoryId'].map(category_mapping)

data.head()
```

```python
data['category_title'].isnull().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
category_counts = data['category_title'].value_counts()

top_categories = category_counts.head(3)

plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue')
plt.title('Top 3 YouTube Categories in Mexico')
plt.xlabel('Category')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.show()
```
